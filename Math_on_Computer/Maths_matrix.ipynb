```python
import numpy as np
import cmath
```

```python
# (AB)* = A*B*
# A*Bconjugate = Aconjugate*Bconjugate
```

```python
i = cmath.sqrt(-1)
a = np.matrix([[-i, 1],[1, i]])
a
```

```python
b = np.matrix([[3, 2+i], [4, 1]])
astar = a.conjugate()
bstar = b.conjugate()
astar, bstar
c = np.matmul(astar, bstar)
c
```

```python
d = np.matmul(a, b)
dstar = d.conjugate()
dstar
```

```python
#Transpose
#(AB)Transpose = Atranspose.Btranspose
```

```python
atrans = a.transpose()
btrans= b.transpose()

abmul = np.matmul(a,b)

abtransmul = abmul.transpose()

atransbtransmul = np.matmul(btrans,atrans)

abtransmul,atransbtransmul
```

# Trace of Matrix

```python
A = np.matrix([[1, 0, 0],[0, 2, 0],[0, 0, 3]])
B = np.matrix([[2, 5, 6], [7, 8, 10], [0, 1, 2]])
```

```python
trace1 = 0
for i in range(len(A)):
  for j in range(len(A)):
    if i == j:
      trace1 = trace1 + A[i, j]
trace2 = 0
for i in range(len(B)):
  for j in range(len(B)):
    if i == j:
      trace2 = trace2 + B[i, j]
tracesum = trace1 + trace2
tracesum
```

```python
C=A+B
trace3 = 0
for i in range(len(C)):
  for j in range(len(C)):
    if i == j:
      trace3 = trace3 + C[i, j]
trace3
```

```python
# Inverse of a matrix
```

```python
X = np.matrix([[1, 2, 3], [4, 5, 6], [7, 2, 9]])
#inverse
Y = np.linalg.inv(X)
```

```python
# Adjoint matrix
np.linalg.det(X)*Y
```

```python
# Determinant of a matrix
np.linalg.det(X)
```

```python
# Rank of a matrix
```

```python
rank = np.linalg.matrix_rank(X)
rank
```
