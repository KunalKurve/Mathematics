```python
from sympy import limit,sqrt,pprint
from sympy.abc import x
```

```python
#1.
exp=((sqrt(2*x+22)-4)/(x+3))
result=limit(exp,x,-3)
print("Limit of")
pprint(exp)
print("at x=-3 is")
print(result)
```

```python
#2.
exp=((2*x*x-17*x+8)/(8-x))
result=limit(exp,x,8)
print("Limit of")
pprint(exp)
print("at x=8 is")
print(result)
```

```python
#3.
exp=((x*x-4*x-21)/(3*x*x-17*x-28))
result=limit(exp,x,7)
print("Limit of")
pprint(exp)
print("at x=7 is")
print(result)
```

```python
#4.
exp=((sqrt(x)-2)/(x-4))
result=limit(exp,x,4)
print("Limit of")
pprint(exp)
print("at x=4 is")
print(result)
```

```python
#5.
exp=((6+4*x)/(x*x+1))
result=limit(exp,x,-3)
print("Limit of")
pprint(exp)
print("at x=-3 is")
print(result)
```
