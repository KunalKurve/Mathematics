```python
from sympy import limit,sqrt,pprint
from sympy.abc import x
```

```python
exp=sqrt(x+10)
result=limit(exp,x,6)
print("Limit of")
pprint(exp)
print("at x=6 is",result)
```

```python
exp=(1+x+x*x)
result=limit(exp,x,0)
print("Limit of")
pprint(exp)
print("at x=0 is",result)
```

```python
exp=(x**4+x**2+1)
r2=sqrt(2)
result=limit(exp,x,r2)
print("Limit of")
pprint(exp)
print("at x=2**1/2 is")
print(result)
```

```python
exp=(x**4+x**2+1)
result=limit(exp,x,sqrt(2))
print("Limit of")
pprint(exp)
print("at x=2**1/2 is")
print(result)
```

```python
exp=((x*x-3*x+2)/(x-2))
result=limit(exp,x,2)
print("Limit of")
pprint(exp)
print("at x=2 is")
print(result)
```

```python
exp=(8-3*x+12*x*x)
result=limit(exp,x,2)
print("Limit of")
pprint(exp)
print("at x=2 is")
print(result)
```

```python
exp=((x*x-25)/(x*x+2*x-15))
result=limit(exp,x,5)
print("Limit of")
pprint(exp)
print("at x=5 is")
print(result)
```

```python
n1=sqrt(x+9)
exp=(x/(3-n1))
result=limit(exp,x,0)
print("Limit of")
pprint(exp)
print("at x=0 is")
print(result)
```

```python
exp=(x/(3-sqrt(x+9)))
result=limit(exp,x,0)
print("Limit of")
pprint(exp)
print("at x=2 is")
print(result)
```
